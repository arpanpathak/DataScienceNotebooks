## Neural Networks and Perceptrons

> A neural network is a collection of units called perceptrons, which mimics a single neuron of the human brain.

> A perceptron is a cell that takes an input vector $ \vec{x} $ and multiplies it with a list of weights $ \vec{w} $, where each weight corresponds to a connection joining the previous node to the current one (just like axons).

> The perceptron applies a linear combination of all the weights, i.e., a dot product, and feeds this result $ z $ to a non-linear activation function, which produces the output $ y $.

###  Linear Combination(Dot product) and firing activation sense

The dot product of the input vector $ \vec{x} $ and weights $ \vec{w} $ can be represented as:

$$
z = \sum_{i=1}^{n} w_i x_i
$$

Then this single percepton applied a non linear activation functon to produce an output.

y = $ \sigma({z}) $ 




## Perceptron Representation

> The perceptron is a foundational model in artificial neural networks, introduced by Frank Rosenblatt in the late 1950s. It consists of a single layer of neurons that linearly combines inputs through weighted sums and applies an activation function to produce an output. While it can effectively classify linearly separable data, its limitations led to the development of more complex architectures like multi-layer perceptrons, which can handle non-linear relationships. The perceptron is significant as it marked a shift towards machine learning approaches that mimic human cognitive processes.

The output of a perceptron can be represented as:

$$
y = \sigma(W x + b)
$$

- $W$: Weight matrix of shape $(m \times n)$
- $x$: Input vector of shape $(n \times 1)$
- $b$: Bias vector of shape $(m \times 1)$
- $\sigma $: Activation function
- $(y)$: Output vector of shape $(m \times 1)$

## Example Matrix Representation of a Neural Network
We could represent the above relation in terms of matrices of weights and inputs to make our calculation easier. Linear algebra comes to the rescue! These operations can be computed by GPU or TPU in parallel.

### 1. Input Layer to First Hidden Layer

- **Input Vector:**
  $$
  \vec{x} = \begin{bmatrix}
  x_1 \\
  x_2 \\
  x_3 \\
  \end{bmatrix}
  $$
  (e.g., $\vec{x} = \begin{bmatrix} 0.5 \\ 0.2 \\ 0.1 \end{bmatrix}$)

- **Weight Matrix:**
  $$
  \mathbf{W}^{(1)} = \begin{bmatrix}
  w_{11} & w_{12} & w_{13} \\
  w_{21} & w_{22} & w_{23}
  \end{bmatrix}
  $$
  (e.g., $\mathbf{W}^{(1)} = \begin{bmatrix} 0.4 & 0.6 & 0.8 \\ 0.1 & 0.3 & 0.5 \end{bmatrix}$)

- **Bias Vector:**
  $$
  \vec{b}^{(1)} = \begin{bmatrix}
  b_1 \\
  b_2
  \end{bmatrix}
  $$
  (e.g., $\vec{b}^{(1)} = \begin{bmatrix} 0.2 \\ 0.1 \end{bmatrix}$)

- **Output of First Hidden Layer:**
  $$
  \vec{h}^{(1)} = \sigma(\mathbf{W}^{(1)} \vec{x} + \vec{b}^{(1)})
  $$

### 2. First Hidden Layer to Second Hidden Layer

- **Weight Matrix:**
  $$
  \mathbf{W}^{(2)} = \begin{bmatrix}
  w_{11}^{(2)} & w_{12}^{(2)} \\
  w_{21}^{(2)} & w_{22}^{(2)}
  \end{bmatrix}
  $$
  (e.g., $\mathbf{W}^{(2)} = \begin{bmatrix} 0.3 & 0.7 \\ 0.2 & 0.4 \end{bmatrix}$)

- **Bias Vector:**
  $$
  \vec{b}^{(2)} = \begin{bmatrix}
  b_1^{(2)} \\
  b_2^{(2)}
  \end{bmatrix}
  $$
  (e.g., $\vec{b}^{(2)} = \begin{bmatrix} 0.1 \\ 0.2 \end{bmatrix}$)

- **Output of Second Hidden Layer:**
  $$
  \vec{h}^{(2)} = \sigma(\mathbf{W}^{(2)} \vec{h}^{(1)} + \vec{b}^{(2)})
  $$

### 3. Last Hidden Layer to Output Layer

- **Weight Matrix:**
  $$
  \mathbf{W}^{(L)} = \begin{bmatrix}
  w_{1} \\
  w_{2}
  \end{bmatrix}
  $$
  (e.g., $\mathbf{W}^{(L)} = \begin{bmatrix} 0.6 & 0.9 \end{bmatrix}$)

- **Bias Vector:**
  $$
  \vec{b}^{(L)} = \begin{bmatrix}
  b^{(L)}
  \end{bmatrix}
  $$
  (e.g., $\vec{b}^{(L)} = \begin{bmatrix} 0.3 \end{bmatrix}$)

- **Final Output:**
  $$
  \vec{y} = \sigma(\mathbf{W}^{(L)} \vec{h}^{(L-1)} + \vec{b}^{(L)})
  $$

## Vectors and Matrices
In neural networks, we often represent data in vector and matrix forms.

### Vectors
A **vector** is a one-dimensional array that can represent data points, weights, and biases.

$$
x = \begin{pmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{pmatrix}
$$

### Matrices
A **matrix** is a two-dimensional array that can represent multiple data points or transformations.

$$
W = \begin{pmatrix} w_{11} & w_{12} & \ldots & w_{1m} \\ w_{21} & w_{22} & \ldots & w_{2m} \\ \vdots & \vdots & \ddots & \vdots \\ w_{n1} & w_{n2} & \ldots & w_{nm} \end{pmatrix}
$$

## Dot Product
The dot product is used to compute the weighted sum of inputs in a layer:

$$
z = W \cdot x + b
$$

Where:
- \( W \) is the weight matrix
- \( x \) is the input vector
- \( b \) is the bias vector

## Activation Function
The output from the dot product is typically passed through an activation function \( f \):

$$
a = f(z)
$$

Common activation functions include:
- ReLU: $f(z) = \max(0, z)$
- Sigmoid: $f(z) = \frac{1}{1 + e^{-z}}$
- Softmax: $f(z)_i = \frac{e^{z_i}}{\sum_{j} e^{z_j}}$



```python
# Visualize activation functions
# You can skip this part and directly move to loss functions

import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def softmax(z):
    exp_z = np.exp(z - np.max(z))  # For numerical stability
    return exp_z / exp_z.sum(axis=0)

# Create an input range
z = np.linspace(-10, 10, 100)

# Calculate activation values
relu_values = relu(z)
sigmoid_values = sigmoid(z)

# For softmax, we'll create a 2D input
z_softmax = np.vstack([z, z]).T  # Two-dimensional input for demonstration
softmax_values = softmax(z_softmax)

# Plot the activation functions
plt.figure(figsize=(15, 5))

# ReLU
plt.subplot(1, 3, 1)
plt.plot(z, relu_values, label='ReLU', color='blue')
plt.title('ReLU Activation Function')
plt.xlabel('Input (z)')
plt.ylabel('Output')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')

# Sigmoid
plt.subplot(1, 3, 2)
plt.plot(z, sigmoid_values, label='Sigmoid', color='orange')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input (z)')
plt.ylabel('Output')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')

# Softmax
plt.subplot(1, 3, 3)
plt.plot(z, softmax_values[:, 0], label='Softmax (z1)', color='green')
plt.plot(z, softmax_values[:, 1], label='Softmax (z2)', color='red')
plt.title('Softmax Activation Function')
plt.xlabel('Input (z)')
plt.ylabel('Output')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

## <span style="color:blue;">Loss Function</span>

The loss function quantifies the difference between the predicted output and the true output, guiding the optimization of the model during training. The choice of loss function depends on the specific problem being addressed, particularly the nature of the output variable.

### <span style="color:green;">Mean Squared Error (MSE)</span>

For regression tasks, where the output is continuous, the Mean Squared Error (MSE) is commonly used:

$$
\mathcal{L}_{\text{MSE}} = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i)^2
$$

Where:
- $y_i$ is the true label for the $i$-th sample.
- $\hat{y}_i$ is the predicted label for the $i$-th sample.
- $m$ is the number of samples.

MSE is sensitive to outliers because it squares the error, giving higher weight to larger discrepancies. It is a popular choice for regression problems due to its simplicity and effectiveness.

### <span style="color:orange;">Cross-Entropy Loss</span>

For classification tasks, especially with categorical outputs, Cross-Entropy Loss is preferred:

$$
\mathcal{L}_{\text{CE}} = -\frac{1}{m} \sum_{i=1}^{m} \sum_{j=1}^{C} y_{ij} \log(\hat{y}_{ij})
$$

Where:
- $C$ is the number of classes.
- $y_{ij}$ is a binary indicator (0 or 1) if class label $j$ is the correct classification for observation $i$.
- $\hat{y}_{ij}$ is the predicted probability that observation $i$ belongs to class $j$.

Cross-Entropy Loss measures the performance of a classification model whose output is a probability value between 0 and 1. It is effective for multi-class problems and encourages the model to assign higher probabilities to the correct classes.

### <span style="color:red;">Other Loss Functions</span>

- **Binary Cross-Entropy:** Used for binary classification tasks, defined as:

  $$
  \mathcal{L}_{\text{BCE}} = -\frac{1}{m} \sum_{i=1}^{m} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
  $$

- **Hinge Loss:** Commonly used for "maximum-margin" classification, primarily with Support Vector Machines (SVM):

  $$
  \mathcal{L}_{\text{hinge}} = \frac{1}{m} \sum_{i=1}^{m} \max(0, 1 - y_i \cdot \hat{y}_i)
  $$

- **Huber Loss:** A combination of MSE and MAE, which is less sensitive to outliers than MSE:

  $$
  \mathcal{L}_{\text{Huber}} = \begin{cases} 
    \frac{1}{2} (y_i - \hat{y}_i)^2 & \text{for } |y_i - \hat{y}_i| \leq \delta \\
    \delta (|y_i - \hat{y}_i| - \frac{1}{2} \delta) & \text{otherwise}
  \end{cases}
  $$

### <span style="color:purple;">Conclusion</span>

Choosing the appropriate loss function is crucial for model performance. For regression problems, MSE is a solid choice, while for classification tasks, Cross-Entropy is preferred. Understanding the characteristics of each loss function helps in tailoring the training process to achieve better results.

# Derivative of Mean Squared Error Loss Function

Now let's see how we can find the derivatives of the loss function. This is very important to train the Neural Network and learn by adjusting weights and biases.

The Mean Squared Error (MSE) loss function is defined as:

$$
\mathcal{L} = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i)^2
$$

Where:
- \( y_i \) is the true label for the \( i \)-th sample.
- \( \hat{y}_i \) is the predicted label for the \( i \)-th sample.
- \( m \) is the number of samples.

To find the derivative of the loss function with respect to the predicted output \( \hat{y} \):

1. **Differentiate MSE:**

$$
\frac{\partial \mathcal{L}}{\partial \hat{y}_i} = \frac{2}{m} (\hat{y}_i - y_i)
$$

This derivative shows how much the predicted output needs to change to reduce the loss, which is crucial for weight updates during backpropagation.

### Derivative of Mean Squared Error Loss Function with Respect to Weights and Bias

The predicted output can be expressed as:

$$
\hat{y}_i = W^T x_i + b
$$

Where:
- \( W \) is the weight vector,
- \( x_i \) is the input vector for the \( i \)-th sample,
- \( b \) is the bias term.

#### Derivative with Respect to Weights \( W \)

To find the derivative of the loss function with respect to the weights \( W \):

Using the chain rule, we can express it as:

$$
\frac{\partial \mathcal{L}}{\partial W} = \frac{\partial \mathcal{L}}{\partial \hat{y}_i} \cdot \frac{\partial \hat{y}_i}{\partial z_i} \cdot \frac{\partial z_i}{\partial W}
$$

Where \( z_i = W^T x_i + b \).

The derivative of the predicted output with respect to \( z_i \) is given by the activation function's derivative:

$$
\frac{\partial \hat{y}_i}{\partial z_i} = f'(z_i)
$$

And the derivative of \( z_i \) with respect to \( W \) is:

$$
\frac{\partial z_i}{\partial W} = x_i
$$

Thus, we can express the full derivative as:

$$
\frac{\partial \mathcal{L}}{\partial W} = \frac{\partial \mathcal{L}}{\partial \hat{y}_i} \cdot f'(z_i) \cdot x_i
$$

Substituting this in, we get:

$$
\frac{\partial \mathcal{L}}{\partial W} = \frac{2}{m} (\hat{y}_i - y_i) \cdot f'(z_i) \cdot x_i
$$

This derivative indicates how much the weights need to change to minimize the loss.

#### Derivative with Respect to Bias \( b \)

To find the derivative of the loss function with respect to the bias \( b \):

Using the chain rule, we can express it as:

$$
\frac{\partial \mathcal{L}}{\partial b} = \frac{\partial \mathcal{L}}{\partial \hat{y}_i} \cdot \frac{\partial \hat{y}_i}{\partial z_i} \cdot \frac{\partial z_i}{\partial b}
$$

The derivative of the predicted output with respect to \( z_i \) is:

$$
\frac{\partial \hat{y}_i}{\partial z_i} = f'(z_i)
$$

And the derivative of \( z_i \) with respect to \( b \) is:

$$
\frac{\partial z_i}{\partial b} = 1
$$

Substituting this in, we get:

$$
\frac{\partial \mathcal{L}}{\partial b} = \frac{\partial \mathcal{L}}{\partial \hat{y}_i} \cdot f'(z_i)
$$

Thus, we can express this as:

$$
\frac{\partial \mathcal{L}}{\partial b} = \frac{2}{m} (\hat{y}_i - y_i) \cdot f'(z_i)
$$

This indicates that the gradient of the loss function with respect to the bias \( b \) is also crucial for adjusting the bias during backpropagation.

### Derivative with Non-Linear Activation Functions

When using non-linear activation functions, such as ReLU or Sigmoid, we need to account for their derivatives.

#### For Sigmoid Activation:

The derivative of the Sigmoid function is:

$$
f'(z) = f(z)(1 - f(z))
$$

Thus, we have:

$$
\frac{\partial \hat{y}_i}{\partial z_i} = f'(z_i) = \hat{y}_i(1 - \hat{y}_i)
$$

Substituting this into our earlier equation gives:

$$
\frac{\partial \mathcal{L}}{\partial W} = \frac{2}{m} (\hat{y}_i - y_i) \cdot \hat{y}_i(1 - \hat{y}_i) \cdot x_i
$$

#### For ReLU Activation:

The derivative of the ReLU function is:

$$
f'(z) = \begin{cases} 
1 & \text{if } z > 0 \\
0 & \text{if } z \leq 0 
\end{cases}
$$

Thus, we have:

$$
\frac{\partial \hat{y}_i}{\partial z_i} = f'(z_i) = \begin{cases} 
1 & \text{if } z_i > 0 \\
0 & \text{if } z_i \leq 0 
\end{cases}
$$

Substituting this into our earlier equation gives:

$$
\frac{\partial \mathcal{L}}{\partial W} = \frac{2}{m} (\hat{y}_i - y_i) \cdot f'(z_i) \cdot x_i
$$

### Conclusion

These derivatives are essential for updating the weights and biases of the neural network during backpropagation. They provide the necessary gradients to minimize the loss function effectively.

# Backpropagation

Backpropagation is a key algorithm in training neural networks. It calculates the gradients of the loss function with respect to the weights and biases, enabling the model to learn from errors by updating its parameters in the right direction.

### Gradients

During backpropagation, we compute the gradients for the weights $\vec{W}$ and bias $b$. These gradients indicate how much the loss function $\mathcal{L}$ changes with respect to small changes in the weights and biases, guiding the optimization process.

For weights $\vec{W}$ and bias $b$, the gradients are defined as follows:

$$
\frac{\partial \mathcal{L}}{\partial \vec{W}} = \frac{1}{m} \sum_{i=1}^{m} ( \hat{y}_i - y_i ) \cdot \vec{x}_i^T
$$

$$
\frac{\partial \mathcal{L}}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} ( \hat{y}_i - y_i )
$$

Where:
- $m$ is the number of samples.
- $y_i$ is the true label for the $i$-th sample.
- $\hat{y}_i$ is the predicted label for the $i$-th sample.
- $\vec{x}_i$ is the input vector for the $i$-th sample.

### Updating Weights and Biases

Once we have computed the gradients, we can update the weights and biases using the following update rules, which leverage vector and matrix operations for efficient computation:

1. **Weight Update:**
   The weights are updated by moving in the opposite direction of the gradient, scaled by a learning rate $\eta$:

   $$
   \vec{W}' = \vec{W} - \eta \cdot \frac{\partial \mathcal{L}}{\partial \vec{W}}
   $$

   This can be expressed in matrix form for all samples:

   $$
   \vec{W}' = \vec{W} - \frac{\eta}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i) \cdot \vec{x}_i^T
   $$

   Here, the term $\sum_{i=1}^{m} (\hat{y}_i - y_i) \cdot \vec{x}_i^T$ produces a vector that indicates how each weight should be adjusted based on the input features.

2. **Bias Update:**
   Similarly, the bias is updated as follows:

   $$
   b' = b - \eta \cdot \frac{\partial \mathcal{L}}{\partial b}
   $$

   In matrix form, this becomes:

   $$
   b' = b - \frac{\eta}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i)
   $$

### Conclusion

Through backpropagation, the model learns to minimize the loss function by systematically adjusting its parameters based on the calculated gradients. The efficient use of vector and matrix operations allows for quick updates across all samples, which is essential for training deep neural networks.

# Implementation

> Let's connect everything together. This code implements a simple feedforward neural network to classify the Iris dataset, utilizing a custom DenseLayer class for the layers. The network consists of a hidden layer with ReLU activation and an output layer with sigmoid activation, designed to predict three classes of Iris flowers. The data is preprocessed through normalization and one-hot encoding before being split into training and testing sets. The model is trained using forward and backward propagation for a specified number of epochs, minimizing the loss calculated by the mean squared error. During training, the loss and accuracy are printed at regular intervals, and finally, the decision boundary is visualized along with the loss history.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

# Define a dense layer for the neural network
class DenseLayer:
    def __init__(self, input_dim, output_dim, activation='relu'):
        # Initialize weights with small random values and biases to zero
        self.weights = np.random.randn(output_dim, input_dim) * 0.1
        self.biases = np.zeros((output_dim, 1))
        self.activation = activation
        self.input_data = None
        self.z_values = None

    def forward(self, input_data):
        # Compute the linear transformation: Z = W * X + b
        self.input_data = input_data
        self.z_values = np.dot(self.weights, self.input_data) + self.biases
        return self._activate(self.z_values)

    def _activate(self, z):
        # Apply the chosen activation function
        if self.activation == 'relu':
            return np.maximum(0, z)  # ReLU: f(z) = max(0, z)
        elif self.activation == 'sigmoid':
            return 1 / (1 + np.exp(-z))  # Sigmoid: f(z) = 1 / (1 + exp(-z))
        elif self.activation == 'linear':
            return z  # Linear activation for output layer
        return z

    def _activation_derivative(self, z):
        # Compute the derivative of the activation function
        if self.activation == 'relu':
            return (z > 0).astype(float)  # ReLU derivative
        elif self.activation == 'sigmoid':
            sig = self._activate(z)
            return sig * (1 - sig)  # Sigmoid derivative
        return np.ones_like(z)  # Derivative of linear activation is 1

    def backward(self, dL_dA, learning_rate):
        # Backpropagation: Compute gradients and update weights/biases
        dL_dZ = dL_dA * self._activation_derivative(self.z_values)  # Gradient of loss w.r.t Z
        m = self.input_data.shape[1]

        dL_dW = np.dot(dL_dZ, self.input_data.T) / m  # Gradient of loss w.r.t W: (1/m) * dZ * X^T
        dL_db = np.sum(dL_dZ, axis=1, keepdims=True) / m  # Gradient of loss w.r.t b: (1/m) * sum(dZ)

        self.weights -= learning_rate * dL_dW  # Update weights
        self.biases -= learning_rate * dL_db  # Update biases
        dL_dInput = np.dot(self.weights.T, dL_dZ)  # Gradient of loss w.r.t input
        return dL_dInput

# Define the neural network structure
class NeuralNetwork:
    def __init__(self):
        self.layers = []
        self.loss_history = []

    def add_layer(self, layer):
        self.layers.append(layer)  # Add a new layer to the network

    def forward(self, X):
        # Pass input through all layers
        for layer in self.layers:
            X = layer.forward(X)
        return X

    def backward(self, Y, output, learning_rate):
        # Compute the loss gradient and backpropagate through the layers
        m = Y.shape[1]
        dL_dOutput = -(Y - output)  # Gradient of loss w.r.t output
        for layer in reversed(self.layers):
            dL_dOutput = layer.backward(dL_dOutput, learning_rate)

    def train(self, X, Y, learning_rate, epochs):
        for epoch in range(epochs):
            output = self.forward(X)  # Forward pass
            loss = np.mean((Y - output) ** 2)  # Mean Squared Error: 
            # L = (1/m) * (Y - )
            self.loss_history.append(loss)

            # Compute accuracy
            predictions = np.argmax(output, axis=0)
            labels = np.argmax(Y, axis=0)
            accuracy = np.mean(predictions == labels)

            self.backward(Y, output, learning_rate)  # Backward pass
            
            # Print loss and accuracy on the same line
            if epoch % 100 == 0:
                print(f"Epoch {epoch}, Loss: {loss:.4f}, Accuracy: {accuracy:.4f}", end='\r')

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for visualization
Y = iris.target.reshape(-1, 1)  # Reshape Y to be a column vector

# One-hot encode the labels
encoder = OneHotEncoder(sparse=False)
Y_encoded = encoder.fit_transform(Y)

# Step 1: Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Train-test split (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y_encoded, test_size=0.2, random_state=42)

# Initialize the neural network
nn = NeuralNetwork()
nn.add_layer(DenseLayer(input_dim=X_train.shape[1], output_dim=20, activation='relu'))  # Hidden layer
nn.add_layer(DenseLayer(input_dim=20, output_dim=20, activation='relu'))  # Hidden layer
nn.add_layer(DenseLayer(input_dim=20, output_dim=3, activation='sigmoid'))  # Output layer

# Train the neural network
learning_rate = 0.03
epochs = 200000
nn.train(X_train.T, Y_train.T, learning_rate, epochs)  # Transpose to match shape

# Plot the loss function
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(nn.loss_history)
plt.title('Loss over Epochs')
plt.xlabel('Epochs (x100)')
plt.ylabel('Loss')
plt.grid()

# Optional: Visualizing the decision boundary
def plot_decision_boundary(X, Y, model):
    # Create a mesh grid for plotting the decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    
    # Compute model output for each point in the mesh grid
    Z = model.forward(np.c_[xx.ravel(), yy.ravel()].T)
    Z = np.argmax(Z, axis=0).reshape(xx.shape)  # Get class predictions

    # Plot decision boundary and training points
    plt.subplot(1, 2, 2)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=np.argmax(Y, axis=1), edgecolors='k')
    plt.title('Decision Boundary')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

# Plot the decision boundary
plot_decision_boundary(X_train, Y_train, nn)

# Show both plots
plt.tight_layout()
plt.show()
```
